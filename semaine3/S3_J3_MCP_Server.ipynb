## Prompt

Template de prompt réutilisable

````python
from dataclasses import dataclass

@dataclass
class Tool:

    name: str
    description: str
````

# Premier serveur MCP

````python
class MCPServer:

    def __init__(self):

        self.tools = {}

    def register_tool(
        self,
        name,
        function,
        description=""
    ):

        self.tools[name] = {
            "function": function,
            "description": description
        }

    def list_tools(self):

        return {
            name: tool["description"]
            for name, tool in self.tools.items()
        }

    def execute(
        self,
        tool_name,
        **kwargs
    ):

        return self.tools[tool_name]["function"](
            **kwargs
        )
````

# Premier outil

````python
def weather(city):

    return f"Météo simulée pour {city}"

server = MCPServer()

server.register_tool(
    "weather",
    weather,
    "Retourne la météo"
)

server.list_tools()

server.execute(
    "weather",
    city="Paris"
)
````

# Ajouter plusieurs outils

````python
def calculate(a, b):

    return a + b


def stock_price(symbol):

    return f"Cours simulé : {symbol}"


server.register_tool(
    "calculate",
    calculate,
    "Additionne deux nombres"
)

server.register_tool(
    "stock_price",
    stock_price,
    "Retourne un cours boursier"
)

server.list_tools()
````

# Découverte dynamique

L'un des intérêts majeurs de MCP :

Le client peut découvrir automatiquement les outils disponibles.

````python
for name, description in server.list_tools().items():

    print(name)
    print(description)
    print()
````

# Construire un client MCP

````python
class MCPClient:

    def __init__(self, server):

        self.server = server

    def discover(self):

        return self.server.list_tools()

    def call(
        self,
        tool_name,
        **kwargs
    ):

        return self.server.execute(
            tool_name,
            **kwargs
        )

client = MCPClient(server)

client.discover()


client.call(
    "weather",
    city="Lyon"
)
````
