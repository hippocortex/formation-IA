# S3 J3 - Construire un serveur MCP en Python

Objectif : comprendre les concepts MCP Server, Tools, Resources et leur exposition.

```python
class MCPServer:
    def __init__(self):
        self.tools = {}

    def register_tool(self, name, func):
        self.tools[name] = func

    def list_tools(self):
        return list(self.tools.keys())

    def execute(self, name, **kwargs):
        return self.tools[name](**kwargs)
```

```python
def weather(city):
    return f"Météo simulée pour {city}"

server = MCPServer()
server.register_tool("weather", weather)

print(server.list_tools())
print(server.execute("weather", city="Paris"))
```
