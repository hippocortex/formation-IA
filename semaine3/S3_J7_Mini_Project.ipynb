# S3 J7 - Mini Projet Multi-Agent

Assembler Planner, Researcher, Writer et un MCP Server.

```python
class Planner:
    def run(self):
        return "Plan"

class Researcher:
    def run(self):
        return "Données"

class Writer:
    def run(self, plan, data):
        return f"Rapport basé sur {plan} et {data}"

planner = Planner()
researcher = Researcher()
writer = Writer()

report = writer.run(planner.run(), researcher.run())
print(report)
```
