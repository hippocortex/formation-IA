
# Semaine 3 – Jour 2  
## Agents IA avec LangChain : Tools & exécution

### Objectifs pédagogiques
- Comprendre le rôle des **tools** dans un agent LangChain
- Créer un agent capable de choisir dynamiquement un outil
- Comprendre le cycle **Reason → Act → Observe**

### Contexte
Un agent n'est pas seulement un LLM :  
il **raisonne**, **choisit une action**, **appelle un outil**, puis **exploite le résultat**.

Aujourd’hui, nous implémentons un **agent LangChain simple** capable :
- d’utiliser un outil mathématique
- ou un outil de recherche simulé



## 1. Installation et imports


```

# Installation (à exécuter si nécessaire)
# !pip install langchain langchain-openai

from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, Tool
from langchain.agents.agent_types import AgentType
```


## 2. Définition des tools

Un **tool** est :
- une fonction Python
- avec un nom clair
- une description explicite (très importante pour le LLM)


```

def calculator(expression: str) -> str:
    """Évalue une expression mathématique simple."""
    try:
        return str(eval(expression))
    except Exception as e:
        return f"Erreur de calcul : {e}"

def fake_search(query: str) -> str:
    return f"Résultat simulé pour la recherche : '{query}'"
```


## 3. Enregistrement des tools dans LangChain


```

tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Utile pour effectuer des calculs mathématiques"
    ),
    Tool(
        name="Search",
        func=fake_search,
        description="Utile pour rechercher des informations générales"
    )
]
```


## 4. Création du LLM


```

llm = ChatOpenAI(
    model="gpt-4o-mini",  # ou autre modèle
    temperature=0
)
```


## 5. Initialisation de l’agent

Nous utilisons ici :
- **ZERO_SHOT_REACT_DESCRIPTION**
- le modèle décide seul quel tool utiliser


```

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```


## 6. Exécution de l’agent


```

agent.run("Combien font 12 * (4 + 6) ?")
```

```

agent.run("Recherche des informations sur le protocole MCP") 
```


## 7. Analyse pédagogique

### Ce que fait l’agent :
1. Analyse la question
2. Choisit un tool
3. Exécute le tool
4. Interprète la réponse

### Points clés à retenir
- La **description des tools** est critique
- L’agent n’exécute jamais de code directement
- Tout passe par une **action contrôlée**
