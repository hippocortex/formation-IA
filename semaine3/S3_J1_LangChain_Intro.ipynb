## 1️⃣ Architecture générale de LangChain

LangChain repose sur 5 piliers :

| Composant | Description |
|------------|-------------|
| **LLMs** | Interface avec les grands modèles de langage (GPT, Claude, Gemini…) |
| **PromptTemplates** | Gabarits pour structurer les prompts |
| **Chains** | Enchaînements logiques de prompts et d’actions |
| **Tools** | Fonctions externes appelables par le modèle (calcul, recherche web…) |
| **Agents** | Entités capables de choisir quels outils utiliser selon le contexte |

```text
Utilisateur → Agent → (Prompt + Tools + Memory) → LLM → Réponse
```

## 3️⃣ Créer un prompt et une chaîne simple

Une *chain* (chaîne) combine un modèle, un prompt et éventuellement une logique d’entrée/sortie.

````python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain

llm = ChatOpenAI(model='gpt-3.5-turbo', temperature=0.7)
prompt = ChatPromptTemplate.from_template('Traduis en français : {texte}')

chain = LLMChain(llm=llm, prompt=prompt)

resultat = chain.run(texte='Hello, my name is LangChain!')
print(resultat)
````

## 4️⃣ Ajouter un outil (Tool) et un agent

Un **Tool** est une fonction Python accessible à l’agent. Par exemple, une calculatrice.

````python
from langchain.tools import tool

@tool
def calculer(expression: str) -> str:
    """Évalue une expression mathématique simple."""
    try:
        return str(eval(expression))
    except Exception as e:
        return f"Erreur : {e}"

print(calculer('2 + 3 * 5'))
````

Créons maintenant un agent capable de **choisir quand utiliser cet outil** pour répondre :

````python
from langchain.agents import initialize_agent, AgentType

tools = [calculer]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

response = agent.run('Quel est le carré de 12 plus 5 ?')
print(response)
````

## 5️⃣ Comparaison : Chain vs Agent

| Élément | Chain | Agent |
|----------|--------|--------|
| Logique | Séquentielle | Dynamique (choisit ses actions) |
| Cas d’usage | Traduction, résumé, génération | Chatbots, recherche, exécution multi-étapes |
| Complexité | Faible | Moyenne à élevée |
