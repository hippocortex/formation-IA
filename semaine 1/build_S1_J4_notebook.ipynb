## 🎯 Few-shot Prompting
Le *few-shot prompting* consiste à fournir au modèle quelques exemples d’entrées et de sorties avant de poser la vraie question.

Exemple :
```
Traduis en anglais :
Français : Bonjour -> Anglais : Hello
Français : Merci -> Anglais : Thank you
Français : Au revoir -> Anglais :
```
Le modèle infère le schéma attendu.

````python
def few_shot_demo(prompt, examples):
    print("=== Contexte ===")
    for ex in examples:
        print(f"Entrée : {ex['input']} -> Sortie : {ex['output']}")
    print("\n=== Question ===")
    print(prompt)
    print("\n=== Réponse simulée ===")
    return "Réponse probable basée sur le schéma appris"

examples = [
    {"input": "Bonjour", "output": "Hello"},
    {"input": "Merci", "output": "Thank you"}
]

result = few_shot_demo("Comment traduire 'Au revoir' ?", examples)
print(result)
````

## 🧩 Chain-of-Thought (CoT)
Le *Chain-of-Thought* pousse le modèle à **raisonner étape par étape** avant de donner une réponse.

Exemple :
```
Question : Si un train part à 100 km/h et un autre à 120 km/h, quand se rencontrent-ils ?
Réponse : Réfléchissons étape par étape...
```

````python
def chain_of_thought_simulation(question):
    print(f"Question : {question}")
    print("Raisonnement :")
    print("1. Identifier les vitesses des deux trains.")
    print("2. Calculer la distance relative.")
    print("3. Trouver le temps où les positions s'égalisent.")
    print("\nRéponse finale : 2 heures.")
    
chain_of_thought_simulation("Deux trains partent à 100 km/h et 120 km/h de points distants de 440 km.")
````

## 🧱 Prompt Patterns
Les *Prompt Patterns* sont des structures réutilisables, comme des patrons de design.

Exemples de patterns :
- **Instruction + contexte + question**
- **Rôle + objectif + contrainte**
- **Tableau d’exemples (few-shot pattern)**

```python
pattern = {
    "role": "Tu es un expert en IA.",
    "objectif": "Expliquer simplement un concept.",
    "contrainte": "Réponse < 100 mots."
}
print(pattern)
```
