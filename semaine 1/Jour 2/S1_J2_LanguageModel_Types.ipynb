## Types de modèles de langage : GPT vs BERT vs T5  

Les modèles de langage modernes reposent sur l’architecture Transformer, mais diffèrent par **leur structure** (auto-régressive, auto-encodage ou encoder-decoder) et leurs **usages cibles**.


## 1. Typologie des modèles

| Type | Exemples | Architecture | Tâches principales | Mode d'entraînement | Forces | Limites |
|------|-----------|--------------|--------------------|---------------------|---------|----------|
| **Auto-régressif** | GPT, GPT-2, GPT-3, GPT-4 | *Decoder-only* | Génération de texte, complétion, chat | Prédire le mot suivant (causal LM) | Excellente génération fluide et cohérente | Ne comprend pas bien les trous dans le texte |
| **Auto-encodeur** | BERT, RoBERTa | *Encoder-only* | Compréhension, classification, QA fermée | Reconstruction de tokens masqués | Très bon en compréhension fine | Pas conçu pour générer |
| **Encoder-Decoder** | T5, BART, FLAN-T5 | *Seq2Seq* | Traduction, résumé, question-réponse, génération guidée | Encode + décode séquentiellement | Polyvalent, bonne compréhension et génération | Plus coûteux à entraîner |


## 2. Schéma simplifié

```text
GPT (Decoder-only):        [Texte Entrée] ---> [Decoder] ---> [Texte Généré]

BERT (Encoder-only):       [Texte Entrée] ---> [Encoder] ---> [Représentation]

T5 (Encoder-Decoder):      [Texte Entrée] ---> [Encoder] ---> [Decoder] ---> [Texte Sortie]
```


## 3. Synthèse visuelle

- 🗣️ **GPT** = Générateur pur  
- 🧩 **BERT** = Compreneur  
- 🔁 **T5** = Traducteur généraliste  

Souviens-toi :  
> 👉 “GPT parle, BERT comprend, T5 fait les deux.”
