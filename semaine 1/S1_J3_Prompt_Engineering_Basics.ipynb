# 🧠 Semaine 1 — Jour 3 : Prompt Engineering — Les bases

Ce notebook t’initie aux fondamentaux du **Prompt Engineering**, c’est-à-dire l’art de formuler les instructions pour obtenir des résultats précis et utiles des modèles de langage (LLM).

## Objectifs
- Comprendre ce qu’est un *prompt* et son influence sur la sortie d’un modèle.
- Apprendre à structurer un prompt clair et efficace.
- Tester plusieurs approches : zéro-shot, few-shot, et chain-of-thought.
- Observer l’impact de la formulation et du contexte sur la réponse du modèle.

## 🧩 1. Qu’est-ce qu’un prompt ?

Un **prompt** est une *entrée textuelle* fournie à un modèle de langage. Il décrit la tâche que tu veux accomplir.

Exemples :
- “Traduis ce texte en anglais.”
- “Écris une courte histoire sur un renard et un robot.”
- “Résume cet article en trois phrases.”

Le modèle produit une **réponse (output)** qui dépend fortement de la **formulation** du prompt.

## 🧠 2. Les bonnes pratiques de base

Quelques règles d’or :
1. **Sois clair et explicite** : indique le rôle du modèle, le ton attendu, le format de la réponse.
2. **Donne du contexte** : plus tu fournis d’informations pertinentes, mieux le modèle comprend la tâche.
3. **Structure ta demande** : sépare les étapes ou contraintes dans ton prompt.
4. **Teste et ajuste** : reformule jusqu’à obtenir une sortie stable.

👉 Exemple :

````python
from openai import OpenAI
client = OpenAI()

prompt = "Tu es un professeur de science. Explique le concept de gravité à un enfant de 8 ans en trois phrases simples."

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)

print(response.choices[0].message.content)
````

## 🧩 3. Les types de prompting

| Type | Description | Exemple |
|------|--------------|----------|
| **Zéro-shot** | Le modèle reçoit uniquement la tâche. | “Traduis ce texte en anglais.” |
| **Few-shot** | Tu fournis quelques exemples pour guider le modèle. | “Voici des exemples de traduction...” |
| **Chain-of-thought (CoT)** | Tu demandes au modèle de *raisonner étape par étape*. | “Explique ton raisonnement avant de répondre.” |

### 🔹 Zéro-shot Prompting

````python
prompt = "Explique pourquoi le ciel est bleu."

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
````

### 🔹 Few-shot Prompting

````python
prompt = (
    "Traduis en anglais :\n"
    "Chat → Cat\n"
    "Chien → Dog\n"
    "Oiseau →"
)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
````

### 🔹 Chain-of-Thought Prompting

````python
prompt = (
    "Un train quitte Paris à 14h et roule à 120 km/h. Un autre quitte Lyon à 15h et roule à 160 km/h."
    "La distance entre Paris et Lyon est de 460 km. À quelle heure se croisent-ils ?"
    "Explique ton raisonnement étape par étape."
)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
````

## 🔍 4. Comparaison d’approches : influence du rôle et du format

Un même sujet, différents rôles et formats :

````python
roles = [
    "Tu es un historien.",
    "Tu es un humoriste.",
    "Tu es un journaliste scientifique.",
]

for role in roles:
    prompt = f"{role} Explique pourquoi les chats retombent sur leurs pattes."
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    print(f"---\n{role}\n{response.choices[0].message.content}\n")
````

## 🧪 5. Atelier — Optimisation de prompt

Testons l’impact des consignes sur le format de la réponse.

**Tâche :** Résumer ce texte en 3 phrases, puis en format puces, puis en haïku !

````python
texte = """Les abeilles jouent un rôle crucial dans la pollinisation des plantes, ce qui permet la production de fruits et légumes. 
Sans elles, de nombreuses cultures alimentaires disparaîtraient. Leur déclin est un problème écologique majeur."""

prompts = [
    f"Résume ce texte en trois phrases :\n{texte}",
    f"Résume ce texte sous forme de puces :\n{texte}",
    f"Résume ce texte sous forme de haïku :\n{texte}",
]

for p in prompts:
    r = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": p}]
    )
    print("---\n", r.choices[0].message.content, "\n")
````
