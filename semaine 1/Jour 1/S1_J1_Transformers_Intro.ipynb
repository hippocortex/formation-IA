## Mini Démo – Self-Attention simulée avec NumPy

Ce notebook illustre le mécanisme d’**attention** utilisé dans les Transformers, à travers une implémentation simple.


```

import numpy as np

# Pour l'affichage plus clair
np.set_printoptions(precision=3, suppress=True)
```

## 🎯 Exemple de phrase et initialisation

```

# Supposons une phrase simplifiée avec 3 mots : "Je", "mange", "pomme"
tokens = ["Je", "mange", "pomme"]

# Embeddings aléatoires simulés (taille 4)
np.random.seed(42)
embeddings = np.random.rand(len(tokens), 4)

print("🔢 Embeddings simulés :\n", embeddings)
```

## 🧮 Création des matrices Q, K, V

```

# Poids aléatoires pour Q, K, V
W_Q = np.random.rand(4, 4)
W_K = np.random.rand(4, 4)
W_V = np.random.rand(4, 4)

# Calcul des vecteurs Q, K, V
Q = embeddings @ W_Q
K = embeddings @ W_K
V = embeddings @ W_V

print("Q:\n", Q, "\n\nK:\n", K, "\n\nV:\n", V)
```

## ⚡ Calcul du score d’attention

```

# Produit scalaire QK^T
scores = Q @ K.T

# Normalisation par la racine de la dimension (d_k)
d_k = K.shape[-1]
scaled_scores = scores / np.sqrt(d_k)

# Application du softmax
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / exp_x.sum(axis=-1, keepdims=True)

attention_weights = softmax(scaled_scores)
print("🧭 Poids d'attention (chaque ligne = mot qui regarde les autres):\n", attention_weights)
```

## 🧩 Application des poids sur les valeurs V

```

# Combinaison des poids avec les valeurs V
output = attention_weights @ V
print("🎯 Sortie du mécanisme d'attention:\n", output)
```


## 📊 Interprétation
Chaque ligne de `attention_weights` montre **comment un mot "regarde" les autres mots**.  
Par exemple, si `attention_weights[1] = [0.2, 0.7, 0.1]`, cela signifie que **"mange"** accorde :  
- 20% d’attention à "Je"  
- 70% à lui-même  
- 10% à "pomme"  

Ce principe est à la base de la **compréhension contextuelle** des Transformers.
