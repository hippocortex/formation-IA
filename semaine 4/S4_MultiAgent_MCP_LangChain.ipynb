## Mock orchestration (LangChain-style) sans dépendances lourdes
Le code ci-dessous montre une structure d'orchestration où chaque 'agent' est une fonction ou objet qui peut être appelé.

````
# Simple mock orchestrator
class MockAgent:
    def __init__(self, name, role, func):
        self.name = name
        self.role = role
        self.func = func
    def act(self, input):
        return self.func(input)

# Define roles
def researcher_fn(inp):
    return f"findings about {inp}"

def planner_fn(inp):
    return f"plan for {inp}"

def executor_fn(inp):
    return f"executed {inp}"

researcher = MockAgent('researcher','researcher', researcher_fn)
planner = MockAgent('planner','planner', planner_fn)
executor = MockAgent('executor','executor', executor_fn)

# Orchestrator
def orchestrate(user_req):
    r = researcher.act(user_req)
    p = planner.act(r)
    e = executor.act(p)
    return {'researcher':r, 'planner':p, 'executor':e}

print(orchestrate('organiser demo'))
````

## Logging & visualisation
Utilise networkx/matplotlib pour tracer le flux d'interactions entre agents (cf. notebook Python).

````
import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()
G.add_edge('user','researcher')
G.add_edge('researcher','planner')
G.add_edge('planner','executor')
G.add_edge('executor','user')
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgreen')
plt.show()
````
