# S4 - Jour 2
## Construire un framework d'agents

Cette première partie explique **pourquoi** les frameworks d'agents possèdent tous des composants similaires. Nous nous concentrons sur les décisions d'architecture, pas encore sur l'implémentation.

## Objectifs

- Comprendre les responsabilités d'un framework
- Distinguer Conversation, Memory, Context et State
- Préparer l'implémentation du mini framework

## Pourquoi un framework ?

Les premiers agents sont simples. Dès que l'on ajoute plusieurs outils, plusieurs modèles, de la mémoire, du multi-agent ou des workflows, la complexité vient de l'orchestration et non plus du LLM. Les frameworks encapsulent cette complexité.

## Responsabilité unique

Chaque composant doit avoir une raison unique d'évoluer : Message, Conversation, Memory, ToolRegistry, Config et Agent.

## Architecture

```text
Agent
├── Config
├── ToolRegistry
└── Conversation
      └── Messages
            └── Memory
```

## Pourquoi un objet Message ?

Une classe apporte validation, extensibilité, typage, métadonnées et tests plus simples qu'un dictionnaire brut.

## Conversation vs Memory

Conversation = historique d'une session. Memory = informations persistantes réutilisables au-delà d'une session.

## State, Context, Memory

Le State décrit le workflow, le Context correspond aux informations envoyées au modèle, la Memory conserve les connaissances persistantes.

## Comparaison des frameworks

| Notre concept | LangGraph | OpenAI Agents SDK |
|---|---|---|
| Conversation | Messages du State | Session |
| Memory | State Store | Memory |
| Agent | Node/Agent | Agent |

## Notes d'architecte

Évitez un Agent qui concentre toute la logique. Les responsabilités séparées rendent le système testable et évolutif.

## Notes marché

En 2026, LangGraph, l'OpenAI Agents SDK et MCP structurent fortement l'écosystème des agents IA.
