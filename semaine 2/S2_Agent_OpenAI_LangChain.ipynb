## Mock mode — utiliser LangChain patterns sans Appels API
Le code ci-dessous illustre comment structurer un 'Agent' en LangChain de manière mockée.

````
# Mock-like LangChain structure (pseudocode without requiring actual langchain API calls)
class MockTool:
    def __init__(self, name, func):
        self.name = name
        self.func = func
    def run(self, *args, **kwargs):
        return self.func(*args, **kwargs)

# Define mock tools
def mock_search(q):
    return f"Results for {q} (mock)"

def mock_weather(city):
    return {'city':city,'temp_c':15}

search_tool = MockTool('search', mock_search)
weather_tool = MockTool('get_weather', mock_weather)

# Simple orchestrator
def simple_agent_run(user_input):
    if 'météo' in user_input.lower():
        return weather_tool.run(user_input.split()[-1])
    return search_tool.run(user_input)

print('Mock LangChain-style agent ready')
print(simple_agent_run('Quelle est la météo à Lyon'))
````
