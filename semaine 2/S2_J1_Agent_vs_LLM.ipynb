## 1. Introduction : LLM vs Agent IA

### 🔹 LLM (Large Language Model)
Un LLM (comme GPT-4 ou Mistral) est un **modèle statistique** entraîné pour prédire le mot suivant dans une séquence de texte. Il ne dispose **ni de mémoire**, **ni d’objectif**, **ni de capacité d’action directe** sur le monde extérieur.

➡️ Il agit uniquement en fonction du **prompt** qu’on lui fournit.

**Exemples d’usage LLM seul :**
- Rédaction de texte
- Résumé ou traduction
- Chatbot sans état ni logique d’action

### 🔹 Agent IA
Un agent est un **système construit autour d’un LLM**, auquel on ajoute :
- un **objectif** à atteindre (goal)
- une **boucle de raisonnement** (planification, réflexion)
- des **outils externes** (API, bases de données, fonctions)
- parfois une **mémoire** (contexte persistant)

➡️ Il peut **raisonner, agir et s’auto-corriger**.

**Exemples d’usage agent :**
- Assistant personnel (lecture d’agenda, météo, tâches)
- Analyste de données automatique
- Chatbot connecté à un CRM ou à une API tierce
- Agent multi-étapes pour RAG (Retrieval Augmented Generation)

## 2. Architecture d’un agent IA

Un agent IA suit généralement une **boucle de raisonnement-action** :

```text
Entrée utilisateur → Raisonnement (LLM) → Choix d’une action → Exécution → Observation → Réévaluation
```

### 🔧 Composants typiques
- **LLM Core** : le moteur de raisonnement
- **Toolset** : fonctions que l’agent peut invoquer
- **Memory** : historique ou contexte long
- **Controller / Orchestrator** : gère la logique de décision

➡️ Cette architecture est inspirée des agents cognitifs (planification + perception + action).

## 3. Cas d’usage : LLM seul vs Agent

| Situation | LLM seul | Agent IA |
|------------|-----------|-----------|
| Générer un texte | ✅ | ✅ |
| Résumer un document | ✅ | ✅ |
| Interroger une base externe | ❌ | ✅ |
| Exécuter une action (envoyer un mail, récupérer une donnée) | ❌ | ✅ |
| Résoudre une tâche complexe multi-étapes | ⚠️ | ✅ |

**Conclusion :** un agent est une orchestration autour du LLM qui lui donne du pouvoir d’action.

````python
# 4. Démonstration Python simple : LLM vs Agent minimal

from typing import Callable

# Simulation d'un LLM simple (mock)
def simple_llm(prompt: str) -> str:
    if "météo" in prompt.lower():
        return "Je ne peux pas accéder à la météo, je suis un modèle statique."
    return f"Réponse textuelle basée sur le prompt: {prompt[:50]}..."

# Agent simple avec capacité d'action (appel d'une fonction)
def get_weather(city: str) -> str:
    return f"☀️ Il fait 24°C et ensoleillé à {city}."

def agent(prompt: str, tools: dict[str, Callable]):
    if "météo" in prompt.lower():
        city = prompt.split()[-1]
        return tools['get_weather'](city)
    else:
        return simple_llm(prompt)

tools = {"get_weather": get_weather}

print("LLM simple:", simple_llm("Quelle est la météo à Paris ?"))
print("Agent IA:", agent("Quelle est la météo à Paris ?", tools))
````

## 6. Discussion : Limites et perspectives

**Limites des LLM :**
- Pas de mémoire persistante
- Pas d’accès à des données réelles
- Peu explicables

**Limites des agents :**
- Complexité de conception et débogage
- Risque d’erreurs d’action (hallucinations fonctionnelles)
- Dépendance aux outils externes et au coût d’appel API

**Perspectives :**
- Agents multi-modaux (texte, image, code)
- Coordination entre plusieurs agents (MCP, CrewAI, AutoGen)
- Mémoire vectorielle et apprentissage incrémental
