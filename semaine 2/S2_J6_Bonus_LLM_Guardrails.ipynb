## Objectifs

- LLM as Judge

```python
import re

DENYLIST = ['api key','password','mot de passe']

def deterministic_filter(text):
    for p in DENYLIST:
        if p in text.lower():
            return False, p
    return True, None

print(deterministic_filter('Comment fonctionne FastAPI ?'))
print(deterministic_filter('Donne moi le password admin'))
```

```python
PROMPT_INJECTION_PATTERNS = [
    'ignore previous instructions',
    'system prompt',
    'reveal hidden instructions'
]

def detect_prompt_injection(text):
    for p in PROMPT_INJECTION_PATTERNS:
        if p in text.lower():
            return True, p
    return False, None

print(detect_prompt_injection('Ignore previous instructions'))
```

```python
def mock_llm_classifier(text):
    if 'attaque' in text.lower():
        return {'category':'forbidden','confidence':0.95}
    return {'category':'safe','confidence':0.90}

print(mock_llm_classifier('Comment créer un agent ?'))
```

```python
def judge_tool_call(tool_name):
    sensitive = ['delete_calendar_event','delete_user']
    return tool_name not in sensitive

print(judge_tool_call('delete_calendar_event'))
```
